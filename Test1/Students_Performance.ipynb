```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('StudentsPerformance.csv')
```

```python
df_copy = df.drop(['lunch','test preparation course'],axis=1,inplace=False)
df_copy.head(3)
```

```python
df['parental level of education'] = df['parental level of education'].fillna('default')
```

```python
df.head(7)
```

```python
df['race/ethnicity'] = df['race/ethnicity'].map({
    'group A' : 'Asian',
    'group B' : 'African',
    'group C' : 'Afro-Asian',
    'group D' :'American'
})
```

```python
df.head(7)
```

```python
sns.countplot(data=df,hue='gender',palette='Set1',x='test preparation course')
```

```python
sns.countplot(data=df,x='race/ethnicity',hue='gender')
```

```python
interval = (0,40,50,60,75)
category = ['Fail','2nd Class','Ist Class','Distinction']
```

```python
df['Marks_Cat'] = pd.cut(df.mathscore,interval,labels=category)
```

```python
df.head()
```

```python
sns.countplot(data=df,x='Marks_Cat',palette='Set2',hue='gender')
```

```python
df['Marks_Cat'] = pd.cut(df['reading score'],interval,labels=category)
```

```python
df.head(7)
```

```python
ax = sns.countplot(data=df,x='Marks_Cat',palette='viridis',hue='gender')
ax.set(title='Reading Score',xlabel='Course',ylabel='Count')
plt.show()
```

```python
df['Marks_Cat'] = pd.cut(df['writing score'],interval,labels=category)
ax = sns.countplot(data = df,hue='gender',palette='Set1',x='Marks_Cat')
ax.set(title='Writing Score',xlabel='Course',ylabel='Count')
plt.show()
```
